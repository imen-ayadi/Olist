```python
from sklearn.model_selection import train_test_split

# Extract feature and target arrays
#X, y = prod.drop('_product_category_name_', axis=1), prod[['_product_category_name_']]
```

```python
prod["_payment_type_"] = prod["_payment_type_"].map({'credit_card': 0, 'boleto': 1, 'voucher': 2, 'debit_card': 3})
```

```python
y = prod._product_category_name_
y
```

```python
col = [2,3,4,5,9,10,11,12,13]
X = prod.iloc[:,col]
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=1)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score

# Définir les paramètres du modèle XGBoost
params = {
    'colsample_bytree': 1.0,
    'learning_rate': 0.1,
    'max_depth': 7,
    'n_estimators': 1000,
    'subsample': 1.0
    
}

# Entraîner le modèle
model = xgb.XGBClassifier(**params)
model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
y_pred = model.predict(X_test)

# Évaluer les performances du modèle
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.model_selection import GridSearchCV
# Définir les paramètres à tester
params = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 500, 1000],
    'subsample': [0.5, 0.75, 1.0],
    'colsample_bytree': [0.5, 0.75, 1.0]
}

# Créer le modèle XGBoost
model = xgb.XGBClassifier()

# Effectuer la recherche de grille
grid = GridSearchCV(model, params, cv=3)
grid.fit(X_train, y_train)

# Afficher les meilleurs paramètres trouvés
print('Best Parameters:', grid.best_params_)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model, X_test, y_test,display_labels=y,cmap=plt.cm.Blues)  
plt.show()
```
